### # 5 rows from readcounts.csv

```python
# Real data from readcounts.csv
count_data_first_5 = count_data.head()
count_data_first_5
```

### Generate 4000 rows of synthetic data  

```python
# Generate synthetic data based on the first 5 rows
synthetic_data = pd.DataFrame()
# Iterate over each column in count_data_first_5
for column in count_data_first_5.columns:
    if column == 'Unnamed: 0':  # Keep the 'Unnamed: 0' column as is
        synthetic_data[column] = np.random.choice(count_data_first_5[column], size=4000, replace=True)
        print('synthetic data generated successfully', ...)
    else:  # Generate random variations for numeric columns
        mean = count_data_first_5[column].mean()
        std = count_data_first_5[column].std()
        synthetic_data[column] = np.random.normal(loc=mean, scale=std, size=4000).astype(int)
        synthetic_data[column] = synthetic_data[column].clip(lower=0)  # Ensure no negative values
```

```python
# Save the synthetic data to a CSV file
synthetic_data.to_csv('../data/synthetic_readcounts.csv', index=False)
print('synthetic data saved successfully', ...)
```
